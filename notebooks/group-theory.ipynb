# Basic Concepts

## Group

Group is an abstract mathematical object. It is defined as a set of elements $G$ with a binary operation $*$ which follows the following set of axioms:

1. Closure $x, y \in G \implies x*y \in G$
2. Inverse $x * x^{-1} = e $
3. Identity $x*e = e*x = x $
4. Associativity $(x*y)*z = x*(y*z)  $

Group theory is the study of symmetry. When some property of mathematical object is required to be unchanged or remain invariant under some action, groups naturally arise. The definition of group is the most basic set of axioms that capture symmetry and allows solving equations. 

Following is group of symmetries of rotation(r) and reflection(s) of a square. These are group elements, and the group is said to be "acting" on a square.

## Homomorphism

Homomorphism is a mapping that tries to find structural similarities between two groups. Take two groups $(G,*)$ and $(H,\diamond)$. Also let $f:G \to H$. Then $f$ defines a homomorphism if $f(x*y) = f(x) \diamond f(y)$ for all $x,y \in G$.

Example: Consider two groups $G:(R,+)$ and $H:(R^+,\times)$. Also consider a mapping $f:G \to H$ defined as $f:x \to e^x$. Then,
$$
f(x+y) = f(x)+f(y)
\implies e^{(x+y)} = e^x \times e^y = e^{(x+y)}
$$
Therefore $f$ is a homomorphism.

## Properties of Homomorphism

1. **Identity to Identity**

    Consider two groups $(G,\times)$ and $(H, \diamond)$. Also $f:G \to H$ and $f(x*y)=f(x) \diamond f(y)$. Then $f:1_G \to 1_H$.
    
2. **Inverse to inverse**

    Similarly, $f:x^{-1}_G \to x^{-1}_H$.


## Kernel

Kernel of a homomorphism $f$ is defined as $\ker(f) = \{x \in G | f(x) = 1_H\}$.

When homomorphism $f$ is not one-to-one, $\ker(f)$ has more than 1 element. Kernel is a measure of how $f$ fails to be one-to-one. 
The kernel is also a subgroup of G i.e. $\ker(f) \leq G$.

## Isomorphism

An isomorphism is a homomorphism that is both one-to-one and onto i.e. bijective. Isomorphism represents identical group structures between two different gorups.

Example:
Consider two groups $G:(R^+, \times)$ and $H:(R,+)$. Also lets define $\log: G \to H$. Then we have,

1. $\log(x \times y) = \log(x) + \log(y)$ i.e. log is a homomorphism.
2. $\log(x) = \log(y) \implies e^{\log(x)} = e^{\log(y)} \implies x=y$ i.e. $\log$ is injective.
3. Range of log is $(-\inf,\inf)$ i.e. $\log$ is surjective.

Thus $\log$ is an isomorphism.

## Congruence Class

Let $r_0,r_1,r_2,r_3,r_4$ be infinite sets of integers that would have remainders 0,1,2,3,4 respectively when divided by 5. Then if you pick any number from two of these sets, the result is always in one of the other remaining set. These five sets are called congruence class, and if we treat these sets as numbers, we have group $Z_5 = \{0,1,2,3,4\}$.

If a and b are in same congruence class, a is congruent to b mod n $a \equiv b(mod n)$, i.e. a and b have same remainder when you divide by $n$.

## Subgroup

If a subset H of G forms a group by itself under the group operation of G (i.e., it satisfies closure
and the existence of inverses), then H is said to be a subgroup of G, and this is denoted H≤G. Trivially, any group is a subgroup of itself, and the single element group {e} is a subgroup of any group. If H ≤ G is not one of these trivial cases we say that it is a proper subgroup and use the notation H<G.

## Cosets

If $G$ is a group, and $H$ is a subgroup of $G$, and $g$ is an element of $G$, then

$gH$ = { $gh$ : $h$ an element of $H$ } is the left coset of $H$ in $G$ with respect to $g$, and

$Hg$ = { $hg$ : $h$ an element of $H$ } is the right coset of $H$ in $G$ with respect to $g$.

A coset is a translation of a subgroup by some element of the group. Cosets are in general are not a subgroup as they may not contain identity of the group. For all abelian groups, left cosets and right cosets are the same.

## Normal Subgroup

Lets partition group $(z, +)$ into subrgoup $5Z$, and cosets $1+5Z$, $2+5Z$, $3+5Z$ and $4+5Z$. These subgroups and cosets completely cover the original group $Z$. Here, the cosets form a group among themselves, and thus $5Z$ is a *normal subgroup*. This group of cosets is called as *quotient group*.

All subgroups of an Abelian group are normal.

Formally,
Let $N \leq G$, then cosets form a group if $g^{-1}ng \in N$ $\forall y \in G, n \in N$. Then N is called the normal subgroup of G i.e. N $\mathrel{\unlhd}$ G.

## Quotient Group

# Group Action

**Intuition**: An action of group G on X is a way of thinking of the elements of G as symmetries that transform an object like a cube or a set.

**Formally**: Let $G$ be a group and $X$ be a set. Let $S(X)$ be a group of all permutations of the set $X$. An action of $G$ on $X$ is then defined as a homomorphism $\phi: G \to S(X)$. 

In other words, an action of $G$ on $X$ is a function $\phi:G\times X\to X$ with the following properties: $\phi(e,x)=x$ for every $x\in X$, and $\phi(g,\phi(h,x))=\phi(gh,x)$ for every $g,h\in G$ and every $x\in X$.

## Orbit Stabilizer Theorem

**Intuition**: Take one of the faces of a cube. It can go to one of six other faces, and once you’ve decided which face it will go to, one of the vertices on the face has four places it can go, and once you’ve decided that you’ve fixed the rotation. So the total number of rotation of the cube is $6\times 4=24$. Here the set of places that a particular face can go to is nothing other than the orbit of that face. If the chosen face maps to itself, then we have four choices. So the stabilizer of our chosen face has size 4. Multiplying those two numbers together gives us 24, which is the total number of ways we can rotate the cube.

**Formally**: Let $G$ be a group that acts on a set $X$, let $x$ be an element of $X$, let $O_x$ be the orbit of $x$ and let $S_x$ be the stabilizer of $x$. Then $|O_x||S_x|=|G|$.

In words, the size of the group is the size of the orbit times the size of the stabilizer.

# Representation Theory

Geometry asks, “Given a geometric object $X$, what is its group of symmetries?” Representation theory reverses the question to “Given a group $G$, what objects $X$ does it act on?” and attempts to answer this question by classifying such $X$ up to isomorphism.

To each group element we associate a matrix that acts in the same way as the element. $\rho:G \to C^{d \times d}$, such that  $\rho(g_2g_1)=\rho(g_2)\rho(g_1)$.
1. Two representations are equivalent if $\rho_1(x) = T^{-1}\rho_2(x)T$
2. A representation is reducible if $T^{-1}\rho(x)T = \begin{bmatrix}
\rho_1(x) & 0 \\ 0 & \rho_2(x)
\end{bmatrix}$

A complete set of inequivalent irreducible representations we denote $R$

Maschke Theorem

Any representation $\rho$ of a compact group reduces essentially uniquely into a direct sum of irreducible representations.

# Harmonic Analysis

Representation theory is needed to generalize harmonic analysis. The fourier transform on a group is defined as:

$f$ is a function defined on the group. instead of weighing by exponential factor, weigh using the representation matrices.

$$
\widehat{f}(\rho) = \sum_{x \in G} f(x)\rho(x) \quad \forall \rho \in R
$$

**Properties:**
1. Linearity: $\widehat{f+g} = \widehat{f}+\widehat{g}$
2. Unitarity: $<f,g> = <\widehat{f},\widehat{g}>$
3. Left-translation: $\widehat{f^z(\rho)} = p(z)\widehat{f}(\rho)$
4. Convolution: $\widehat{f*g}(\rho) = \widehat{f}(\rho)\widehat{g}(\rho)$


$\widehat{f^{(z)}}(\rho) = \widehat{f}(\rho)\rho(z)$ so right-translation preserves the subspaces spanned by the rowsof $\widehat{f}(\rho)$

**Fourier transform intuituion**: If you take the space of all functions and you apply fourier transform, it defines a way to divide that space into subspaces which transform independantly.

Algebraically, The Fourier Transform $F:f \to \widehat{f}$ is an isomorphism
$$
F: CG \to \bigoplus_{\rho \in R}C^{d_\rho \times d_\rho}
$$

## Group Equivariant Convolutional Networks: Taco Cohen

**Equivariance**

A representation space can obtain its structure from other representation spaces to which it is connected. For this to work, the network or layer Φ that maps one representation to another should be structure preserving. For G-spaces this means that Φ has to be equivariant:
$$
\phi(T_gx) = T'_g\phi(x)
$$

Equivariance can be realized in many ways, and in particular the operators T and T' need not be the same. The only requirement for T and T' is that for any two transformations g and h, we have $T(gh) = T(g)T(h)$ (i.e. T is a linear representation of G).

Besides improving statistical efficiency and facilitating geometrical reasoning, equivariance to symmetry transformations constrains the network in a way that can aid generalization.

**Symmetry Groups**

One simple example of a group is the set of 2D integer translations, $Z^2$. Here the group operation (“composition of transformations”) is addition: (n,m) + (p, q) = (n + p,m+ q). The important properties of the convolution, such as equivariance, arise primarily from the group structure.

**Group $p4$**

The group p4 consists of all compositions of translations and rotations by 90 degrees about any center of rotation in a square grid. A convenient parameterization of this group in terms of three integers r, u, v is
$$
g(r, u, v)=\left[ \begin{array}{ccc}{\cos (r \pi / 2)} & {-\sin (r \pi / 2)} & {u} \\ {\sin (r \pi / 2)} & {\cos (r \pi / 2)} & {v} \\ {0} & {0} & {1}\end{array}\right]
$$

where 0 ≤ r < 4 and (u, v) ∈ $Z^2$. The group operation is given by matrix multiplication. The group $p4$ acts on points in $Z^2$ (pixel coordinates) by multiplying the matrix $g(r, u, v)$ by the homogeneous co-ordinate vector $x(u', v')$ of a point $(u', v')$.

**Group $p4m$**

The group p4m consists of all compositions of translations, mirror reflections, and rotations by 90 degrees about any center of rotation in the grid. Like p4, we can parameterize this group by integers:
$$
g(m, r, u, v)=\left[ \begin{array}{ccc}{(-1)^{m} \cos \left(\frac{r \pi}{2}\right)} & {-(-1)^{m} \sin \left(\frac{r \pi}{2}\right)} & {u} \\ {\sin \left(\frac{r \pi}{2}\right)} & {\cos \left(\frac{r \pi}{2}\right)} & {v} \\ {0} & {0} & {1}\end{array}\right]
$$
where m ∈ {0, 1}, 0 ≤ r < 4 and (u, v) ∈ $Z^2$.


**Feature Maps as functions on Groups**

In convolutional networks, At each pixel co- ordinate (p, q) ∈ $Z^2$, the stack of feature maps returns a K-dimensional vector $f(p, q)$, where $K$ denotes the number of channels. We will be concerned with transformations of the feature maps, so we introduce the following notation for a trans- formation g acting on a set of feature maps:
$$
\left[L_{g} f\right](x)=\left[f \circ g^{-1}\right](x)=f\left(g^{-1} x\right)
$$
Computationally, this says that to get the value of the gtransformed feature map $Lgf$ at the point $x$, we need to do a lookup in the original feature map $f$ at the point $g^{−1}x$, which is the unique point that gets mapped to $x$ by $g$. This operator $L_g$ is a concrete instantiation of the transformation operator $T_g$ referenced above in equivariance section, and one may verify that:
$$
L_{g} L_{h}=L_{g h}
$$
If g represents a pure translation $t = (u, v) ∈ Z^2$ then $g^{−1}x$ simply means $x − t$. The inverse on $g$ in equation above ensures that the function is shifted in the positive direction when using a positive translation, and that $L_g$ satisfies the criterion for being a homomorphism even for transformations g and h that do not commute (i.e. $g_h \neq h_g$).

**Involution**
$$
f^{*}(g)=f\left(g^{-1}\right)
$$

For $\mathbb{Z}^{2}$ feature maps the involution is just a point reflection, but for G-feature maps the meaning depends on the structure of $G$. In all cases, $f^{* *}=f$.

**Group Equivariant Convolution**

At each layer $l$, a regular convnet takes as input a stack of feature maps $f : Z^2 → R^{K^l}$
and convolves or correlates it with a set of $K^{l+1}$ filters $ψ^i : Z^2 → R^K$:

$$
\left[f \star \psi^{i}\right](x)=\sum_{y \in \mathbb{Z}^{2}} \sum_{k=1}^{K^{l}} f_{k}(y) \psi_{k}^{i}(y-x)
$$
Using the substitution $y → y + t$, and leaving out the summation over feature maps for clarity, we see that a translation followed by a correlation is the same as a correlation followed by a translation:

$$
\begin{aligned}\left[\left[L_{t} f\right] \star \psi\right](x) &=\sum_{y} f(y-t) \psi(y-x) \\ &=\sum_{y} f(y) \psi(y+t-x) \\ &=\sum_{y} f(y) \psi(y-(x-t)) \\ &=\left[L_{t}[f \star \psi]\right](x) \end{aligned}
$$

By replacing the shift by a more general transformation from some group G:

$$
\begin{aligned}\left[\left[L_{u} f\right] \star \psi\right](g) &=\sum_{h \in G} \sum_{k} f_{k}\left(u^{-1} h\right) \psi\left(g^{-1} h\right) \\ &=\sum_{h \in G} \sum_{k} f(h) \psi\left(g^{-1} u h\right) \\ &=\sum_{h \in G} \sum_{k} f(h) \psi\left(\left(u^{-1} g\right)^{-1} h\right) \\ &=\left[L_{u}[f \star \psi]\right](g) \end{aligned}
$$

For the full G-correlation, both the inputs f and ψ and the output $f \star ψ$ are functions on G.

**Pointwise Nonlinearities**

G-correlation preserves the transformation properties of the previous layer. What about nonlinearities and pooling?

Recall that we think of feature maps as functions on $G$. In this view, applying a nonlinearity $v : R → R$ to a feature map amounts to function composition. We introduce the composition operator:
$$
C_{\nu} f(g)=[\nu \circ f](g)=\nu(f(g))
$$
which acts on functions by post-composing them with $ν$.

Since the left transformation operator $L$ acts by precomposition, $C$ and $L$ commute:
$$
C_{\nu} L_{h} f=\nu \circ\left[f \circ h^{-1}\right]=[\nu \circ f] \circ h^{-1}=L_{h} C_{\nu} f
$$
so the rectified feature map inherits the transformation properties of the previous layer.
